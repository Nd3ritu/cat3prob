# General gender wage gap
On average women earn 16% less than men , that is,  84 cents for every dollar a man makes. Women of color are among the lowest-paid workers in rural area, with rural black and hispanic women making just 56 cents for every dollar that a rural white , non-hispanic men make.Latinas are compensated just 55% of what non-hispanic white men are paid in 2024. Black women are paid 64% of what non-hispanic white men are paid. Native American women are typically paid only 59 cents for every dollar paid to white, non-hispanic men. A 20-year-old woman just starting full time, year-round work stands to lose $407,760 over a 40-year career compared to her male counterpart.

# DATA ANALYSIS

After collecting data ,it is evident that gender pay gap is a pervasive issue across many industries and regions, where women are systematically paid less than men.This disparity exists across different levels of education,experience and job positions. 
Some of the factors that contribute to the following are:
1. Women are more likely to take career breaks or work part-time due to caregiving responsibilities which leads to slower career progression and reduced earning potential.
2. Women are less likely to negotiate their salaries compared to men, often due to social conditioning and fear of backlash. When women do negotiate, they may face negative perceptions and outcomes.
3.Many organizations lack transparent pay structures, making it difficult to identify and address pay disparities.
4. Women are often concentrated in lower-paying industries and job roles, while men dominate higher-paying fields. Stereotypes and social norms influence career choices, leading to gendered career paths.


# CONCLUSIONS
By addressing the gender pay gap, organizations can foster a more equitable and inclusive environment, enhance employee satisfaction and retention and contribute to broader social and economic equality. 
The issue can be addressed in a number of ways which are:
1. Implement transparent pay scales and salary bands within organizations.
2. Conduct regular pay audits to identify and address gender pay disparities.
3. Enforce stronger equal pay laws and policies at national and local levels.
4. Offer training programs to help women develop negotiation skills.
5. Implement flexible work policies,such as remote work, flexible hours and part-time options which would help women balance work and family responsibilities.
6. Provide affordable and accessible childcare options to reduce the burden of childcare on working mothers allowing them to fully participate in the workforce and thus no need for wage reductions.
7. Encourage and support women to pursue leadership roles through targeted initiatives and quotas if necessary. and thus will increase female representation in decision-making positions hence influencing company policies and culture.
